```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import pearsonr
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dropout, MultiHeadAttention, Concatenate, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Nadam

# ğŸ“„ Load dataset
df = pd.read_csv('/content/cleaned_combined_air_quality_data_with_readable_DATE.csv')
df.rename(columns={"PM2.5": "PM25"}, inplace=True)
df['DATE'] = pd.to_datetime(df['DATE'])

# âš™ Feature Engineering
target = 'PM25'
df[f'{target}_rolling_mean'] = df[target].rolling(3).mean()
df[f'{target}_rolling_std'] = df[target].rolling(3).std()
for i in range(1, 4):
    df[f'{target}_lag_{i}'] = df[target].shift(i)
df['day_of_week'] = df['DATE'].dt.dayofweek
df['month'] = df['DATE'].dt.month
df['WD_sin'] = np.sin(np.radians(df['WD']))
df['WD_cos'] = np.cos(np.radians(df['WD']))
df.drop(columns=['WD'], inplace=True)
df.dropna(inplace=True)

# ğŸ”§ Define all features
all_features = ['PM10', 'NO2', 'SO2', 'FIRECOUNT', 'CO',
                'WS', 'AT', 'RH', 'WD_sin', 'WD_cos',
                f'{target}_rolling_mean', f'{target}_rolling_std',
                'PM25_lag_1', 'PM25_lag_2', 'PM25_lag_3',
                'day_of_week', 'month']

# ğŸ”€ Split source and target domain
source_df = df[df['DATE'] < '2021-01-01']
target_df = df[df['DATE'] >= '2021-01-01']
X_target_all = target_df[all_features]
y_target_all = target_df[target]

# ğŸ¯ CorrXGBoost: Ï„ = 0.31, Î³ = 0.015
correlations = np.array([pearsonr(X_target_all[f], y_target_all)[0] for f in all_features])
gbr = GradientBoostingRegressor()
gbr.fit(X_target_all, y_target_all)
importances = gbr.feature_importances_

tau = 0.31
gamma = 0.015
selected_features = [all_features[i] for i in range(len(all_features))
                     if abs(correlations[i]) >= tau or importances[i] >= gamma]
print("âœ… Selected Features:", selected_features)

def prepare_data(df_part):
    X = df_part[selected_features]
    y = df_part[target]
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()
    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y.values.reshape(-1,1))
    X_reshaped = X_scaled.reshape((X_scaled.shape[0],1,X_scaled.shape[1]))
    return X_reshaped, y_scaled, scaler_X, scaler_y

X_source, y_source, _, _ = prepare_data(source_df)
X_target, y_target, scaler_X_target, scaler_y_target = prepare_data(target_df)

X_target_train, X_target_test, y_target_train, y_target_test, dates_train, dates_test = train_test_split(
    X_target, y_target, target_df['DATE'], test_size=0.2, random_state=42
)

target_test_dates = pd.to_datetime(dates_test).reset_index(drop=True)

# ğŸ”· Model
def build_model(input_shape):
    inputs = Input(shape=input_shape)
    lstm_out = LSTM(100, activation='relu', return_sequences=True)(inputs)
    dropout_out = Dropout(0.4)(lstm_out)
    mha_out = MultiHeadAttention(num_heads=4, key_dim=64)(dropout_out, dropout_out)
    concat_out = Concatenate(axis=-1)([lstm_out, mha_out])
    dense_out = Dense(64, activation='relu')(concat_out)
    output = Dense(1)(dense_out)
    model = Model(inputs=inputs, outputs=output)
    return model

# Pretrain

base_model = build_model((X_source.shape[1], X_source.shape[2]))
base_model.compile(optimizer=Nadam(0.0005), loss='mse')

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

print("\nğŸ”· Pretraining on Source Domain...")
base_model.fit(X_source, y_source, epochs=100, batch_size=32, validation_split=0.2,
               callbacks=[early_stopping, lr_scheduler], verbose=1)

base_model.save_weights('/content/pretrained_on_source.weights.h5')

# Fine-tune
tl_model = build_model((X_target_train.shape[1], X_target_train.shape[2]))
tl_model.load_weights('/content/pretrained_on_source.weights.h5')

for layer in tl_model.layers[:3]:
    layer.trainable = False

tl_model.compile(optimizer=Nadam(0.0005), loss='mse')

print("\nğŸ”· Fine-tuning on Target Domain...")
tl_model.fit(X_target_train, y_target_train, epochs=100, batch_size=32, validation_split=0.2,
             callbacks=[early_stopping, lr_scheduler], verbose=1)

# Predict
y_pred = tl_model.predict(X_target_test)
y_test_rescaled = scaler_y_target.inverse_transform(y_target_test.reshape(-1,1)).flatten()
y_pred_rescaled = scaler_y_target.inverse_transform(y_pred.reshape(-1,1)).flatten()

df_plot = pd.DataFrame({
    'Date': target_test_dates,
    'Actual': y_test_rescaled,
    'Predicted': y_pred_rescaled
}).sort_values(by='Date').reset_index(drop=True)

df_plot['Residuals'] = df_plot['Actual'] - df_plot['Predicted']

# Metrics
mae = mean_absolute_error(df_plot['Actual'], df_plot['Predicted'])
mape = np.mean(np.abs((df_plot['Actual'] - df_plot['Predicted']) / df_plot['Actual'])) * 100
rmse = np.sqrt(mean_squared_error(df_plot['Actual'], df_plot['Predicted']))
r2 = r2_score(df_plot['Actual'], df_plot['Predicted'])

print("\nğŸ”· Transfer Learning Model Evaluation on Target Test Set:")
print(f"MAE:  {mae:.4f}")
print(f"MAPE: {mape:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R^2:  {r2:.4f}")
# ğŸ“‰ Plot: Pretraining MSE Loss
pretrain_history = base_model.history.history
pretrain_train_loss = pretrain_history['loss']
pretrain_val_loss = pretrain_history['val_loss']

plt.figure(figsize=(10, 5))
plt.plot(pretrain_train_loss, label='Training Loss', linewidth=2)
plt.plot(pretrain_val_loss, label='Validation Loss', linewidth=2, color='orange')
plt.title('Pretraining: Training vs Validation Loss (MSE)', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("Pretraining_MSE_Loss_Curve.png", dpi=400)
plt.show()

# ğŸ“‰ Plot: Fine-tuning MSE Loss
finetune_history = tl_model.history.history
finetune_train_loss = finetune_history['loss']
finetune_val_loss = finetune_history['val_loss']

plt.figure(figsize=(10, 5))
plt.plot(finetune_train_loss, label='Training Loss', linewidth=2)
plt.plot(finetune_val_loss, label='Validation Loss', linewidth=2, color='orange')
plt.title('Fine-tuning: Training vs Validation Loss (MSE)', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("Finetuning_MSE_Loss_Curve.png", dpi=400)
plt.show()
# ğŸ“ˆ Actual vs Predicted
plt.figure(figsize=(14,6))
plt.plot(df_plot['Date'], df_plot['Actual'], label='Actual PM2.5', linewidth=2)
plt.plot(df_plot['Date'], df_plot['Predicted'], label='Predicted PM2.5', linewidth=2)
plt.title("Actual vs Predicted PM2.5 on Target Test Set", fontsize=16, fontweight='bold')
plt.xlabel("Date", fontsize=14, fontweight='bold')
plt.ylabel("PM2.5", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig("Actual_vs_Predicted_PM2.5.png", dpi=400)
plt.show()

# ğŸ“ˆ Scatter Plot
plt.figure(figsize=(6,6))
plt.scatter(df_plot['Actual'], df_plot['Predicted'], alpha=0.5)
plt.plot([df_plot['Actual'].min(), df_plot['Actual'].max()],
         [df_plot['Actual'].min(), df_plot['Actual'].max()], 'r--', linewidth=2)
plt.title("Predicted vs Actual PM2.5", fontsize=16, fontweight='bold')
plt.xlabel("Actual PM2.5", fontsize=14, fontweight='bold')
plt.ylabel("Predicted PM2.5", fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig("Predicted_vs_Actual_PM2.5_Scatter.png", dpi=400)
plt.show()

# ğŸ“ˆ Residuals over Time
plt.figure(figsize=(12,4))
plt.plot(df_plot['Date'], df_plot['Residuals'], linewidth=2)
plt.title("Residuals of PM2.5 over Target Test Set", fontsize=16, fontweight='bold')
plt.xlabel("Date", fontsize=14, fontweight='bold')
plt.ylabel("Residual PM2.5 (Actual-Predicted)", fontsize=10, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig("Residuals_over_Time_PM2.5.png", dpi=400)
plt.show()

# ğŸ“ˆ Residuals Histogram
plt.figure(figsize=(6,4))
plt.hist(df_plot['Residuals'], bins=30, edgecolor='k')
plt.title("Residuals Distribution of PM2.5", fontsize=16, fontweight='bold')
plt.xlabel("Residual PM2.5 (Actual-Predicted)", fontsize=12, fontweight='bold')
plt.ylabel("Frequency", fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig("Residuals_Histogram_PM2.5.png", dpi=400)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import pearsonr
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dropout, MultiHeadAttention, Concatenate, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Nadam

# Load dataset
df = pd.read_csv('/content/cleaned_combined_air_quality_data_with_readable_DATE.csv')
df.rename(columns={"PM2.5": "PM25"}, inplace=True)
df['DATE'] = pd.to_datetime(df['DATE'])

# Feature Engineering
target = 'PM25'
df[f'{target}_rolling_mean'] = df[target].rolling(3).mean()
df[f'{target}_rolling_std'] = df[target].rolling(3).std()
for i in range(1, 4):
    df[f'{target}_lag_{i}'] = df[target].shift(i)
df['day_of_week'] = df['DATE'].dt.dayofweek
df['month'] = df['DATE'].dt.month
df['WD_sin'] = np.sin(np.radians(df['WD']))
df['WD_cos'] = np.cos(np.radians(df['WD']))
df.drop(columns=['WD'], inplace=True)
df.dropna(inplace=True)

# Define all features
all_features = ['PM10', 'NO2', 'SO2', 'FIRECOUNT', 'CO', 'WS', 'AT', 'RH', 'WD_sin', 'WD_cos',
                f'{target}_rolling_mean', f'{target}_rolling_std', 'PM25_lag_1', 'PM25_lag_2', 'PM25_lag_3',
                'day_of_week', 'month']

# Split source and target domain
source_df = df[df['DATE'] < '2021-01-01']
target_df = df[df['DATE'] >= '2021-01-01']
X_target_all = target_df[all_features]
y_target_all = target_df[target]

# CorrXGBoost selection
correlations = np.array([pearsonr(X_target_all[f], y_target_all)[0] for f in all_features])
gbr = GradientBoostingRegressor()
gbr.fit(X_target_all, y_target_all)
importances = gbr.feature_importances_

tau = 0.31
gamma = 0.015
selected_features = [all_features[i] for i in range(len(all_features))
                     if abs(correlations[i]) >= tau or importances[i] >= gamma]

def prepare_data(df_part):
    X = df_part[selected_features]
    y = df_part[target]
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()
    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y.values.reshape(-1,1))
    X_reshaped = X_scaled.reshape((X_scaled.shape[0],1,X_scaled.shape[1]))
    return X_reshaped, y_scaled, scaler_X, scaler_y

X_source, y_source, _, _ = prepare_data(source_df)
X_target, y_target, scaler_X_target, scaler_y_target = prepare_data(target_df)

X_target_train, X_target_test, y_target_train, y_target_test, dates_train, dates_test = train_test_split(
    X_target, y_target, target_df['DATE'], test_size=0.2, random_state=42
)
target_test_dates = pd.to_datetime(dates_test).reset_index(drop=True)

# Model building
def build_model(input_shape):
    inputs = Input(shape=input_shape)
    lstm_out = LSTM(100, activation='relu', return_sequences=True)(inputs)
    dropout_out = Dropout(0.4)(lstm_out)
    mha_out = MultiHeadAttention(num_heads=4, key_dim=64)(dropout_out, dropout_out)
    concat_out = Concatenate(axis=-1)([lstm_out, mha_out])
    dense_out = Dense(64, activation='relu')(concat_out)
    output = Dense(1)(dense_out)
    model = Model(inputs=inputs, outputs=output)
    return model

# Pretraining
base_model = build_model((X_source.shape[1], X_source.shape[2]))
base_model.compile(optimizer=Nadam(0.0005), loss='mse')

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

pretrain_history = base_model.fit(X_source, y_source, epochs=100, batch_size=32, validation_split=0.2,
                                  callbacks=[early_stopping, lr_scheduler], verbose=1)
base_model.save_weights('/mnt/data/pretrained_on_source.weights.h5')

# Fine-tuning
tl_model = build_model((X_target_train.shape[1], X_target_train.shape[2]))
tl_model.load_weights('/mnt/data/pretrained_on_source.weights.h5')
for layer in tl_model.layers[:3]:
    layer.trainable = False
tl_model.compile(optimizer=Nadam(0.0005), loss='mse')

finetune_history = tl_model.fit(X_target_train, y_target_train, epochs=100, batch_size=32, validation_split=0.2,
                                callbacks=[early_stopping, lr_scheduler], verbose=1)

# Plotting training and validation loss curves
plt.figure(figsize=(10, 5))
plt.plot(pretrain_history.history['loss'], label='Pretrain - Training Loss')
plt.plot(pretrain_history.history['val_loss'], label='Pretrain - Validation Loss')
plt.title('Pretraining: Training vs Validation Loss (MSE)')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/mnt/data/Pretraining_MSE_Loss_Curve.png", dpi=400)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(finetune_history.history['loss'], label='Finetune - Training Loss')
plt.plot(finetune_history.history['val_loss'], label='Finetune - Validation Loss')
plt.title('Finetuning: Training vs Validation Loss (MSE)')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/mnt/data/Finetuning_MSE_Loss_Curve.png", dpi=400)
plt.show()
```
